```python
from pyspark.sql import SparkSession
```

```python
# Initialize Spark session
spark = SparkSession.builder \
    .appName("ReadCSVExample2") \
    .master("local") \
    .getOrCreate()
```

```python
df = spark.read \
.option("header","True") \
.option("inferschema","True") \
.csv("sample.csv")
```

```python
df.show()
```

```python
filtered_df = df.where("Age > 40").select("Age","Gender","Country","state")
groupeddf = filtered_df.groupBy("Country").count()
input("Press Enter")
groupeddf.show()
```
