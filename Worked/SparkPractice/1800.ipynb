```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.functions import col
```

```python
spark = SparkSession.builder \
        .appName("1800") \
        .getOrCreate()
```

```python
df = spark.read.csv("1800.csv", inferSchema=True)
df.show(5)
```

```python
df2 = df.withColumnRenamed("_c0","A") \
            .withColumnRenamed("_c1","B") \
            .withColumnRenamed("_c2","C") \
            .withColumnRenamed("_c3","D")   \
            .withColumnRenamed("_c4","E") \
            .withColumnRenamed("_c5","F") \
            .withColumnRenamed("_c6","G") \
            .withColumnRenamed("_c7","H")
df2.show(1)
```

```python
df3 = df2.filter(col("C")=="TMAX").show()
```

```python
df3 = df2.filter("C = 'TMAX'")
```

```python
df4 = df3.select("A","C","D")
df4.show()
```

```python
df5 = df4.withColumn("D", col("D")* 0.1 * (9.0 / 5.0) + 32.0)
df5.show()
```

```python
df6 = df5.groupBy("A").agg(min("D").alias("Min Temp"))
df6.show()
```
